```python
def classify_bowling_style(style):
    if pd.isna(style):
        return 'Unknown'
    style = style.lower()
    pace_keywords = ['fast', 'medium']
    left_keywords = ['left']
    right_keywords = ['right']
    spin_keywords = ['offbreak', 'legbreak', 'googly', 'orthodox', 'wrist spin']
    is_pace = any(word in style for word in pace_keywords)
    is_spin = any(word in style for word in spin_keywords)
    is_left = any(word in style for word in left_keywords)
    is_right = any(word in style for word in right_keywords)
    if is_pace and is_spin:
        return 'All-rounder (Pace & Spin)'
    elif is_pace:
        if is_left:
            return 'Left-arm Pacer'
        elif is_right:
            return 'Right-arm Pacer'
        else:
            return 'Pacer' 
    elif is_spin:
        return 'Spin'
    else:
        return 'Unknown'
```

```python
df['Bowling Type'] = df['Bowling Style'].apply(classify_bowling_style)
```
