```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_excel('Datasets/T20.xlsx')
df1.head()
```

```python
df1.drop(['Player','HS','BF','Span','NO','0'],axis=1,inplace=True)
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.replace('-',np.nan,inplace=True)
```

```python
df1['SR']=pd.to_numeric(df1['SR'],errors='coerce')
df1['Runs']=pd.to_numeric(df1['Runs'],errors='coerce').astype('Int64')
df1['Avg']=pd.to_numeric(df1['Avg'],errors='coerce')
df1['100']=pd.to_numeric(df1['100'],errors='coerce').astype('Int64')
df1['50']=pd.to_numeric(df1['50'],errors='coerce').astype('Int64')
df1['4s']=pd.to_numeric(df1['4s'],errors='coerce').astype('Int64')
df1['6s']=pd.to_numeric(df1['6s'],errors='coerce').astype('Int64')
```

```python
df1.info()
```

```python
df2=pd.read_excel('Datasets/T20.xlsx',sheet_name='Bowling')
df2.head()
```

```python
df2.drop(['Player','Span','Runs','BBI'],axis=1,inplace=True)
```

```python
df2.info()
```

```python
df2['Wkts']=pd.to_numeric(df2['Wkts'],errors='coerce').astype('Int64')
df2['Innings']=pd.to_numeric(df2['Innings'],errors='coerce').astype('Int64')
df2['Overs']=pd.to_numeric(df2['Overs'],errors='coerce')
df2['Madeins']=pd.to_numeric(df2['Madeins'],errors='coerce').astype('Int64')
df2['Avg']=pd.to_numeric(df2['Avg'],errors='coerce')
df2['Econ']=pd.to_numeric(df2['Econ'],errors='coerce')
df2['SR']=pd.to_numeric(df2['SR'],errors='coerce')
df2['5']=pd.to_numeric(df2['5'],errors='coerce').astype('Int64')
df2['4']=pd.to_numeric(df2['4'],errors='coerce').astype('Int64')
```

```python
df2.replace('-',np.nan,inplace=True)
```

```python
df2.info()
```

```python
df3=pd.read_excel('Datasets/T20.xlsx',sheet_name='Fielding')
df3.head()
```

```python
df3.drop(['Player','Innings','MD','Span','Innings'],axis=1,inplace=True)
```

```python
df3.replace('-',np.nan,inplace=True)
```

```python
df3.info()
```

```python
df1.dropna(inplace=True)
df2.dropna(inplace=True)
df3.dropna(inplace=True)
```

```python
with pd.ExcelWriter("Cleaned_datasets/T20.xlsx") as writer:
    df1.to_excel(writer, sheet_name="Batting",index=False)
    df2.to_excel(writer, sheet_name="Bowling", index=False)
    df3.to_excel(writer, sheet_name="Fielding", index=False)
```
